```python
import seaborn as sns
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition 2.csv')
df.head()
```

```python
dummies = pd.get_dummies(df['Attrition'])
dummies.head(1)
```

```python
df = pd.concat([df, dummies], axis=1) 
df.head(5)
```

```python
from sklearn.preprocessing import normalize 

X = df[['EmployeeCount', 'DailyRate', 'DistanceFromHome', 'TotalWorkingYears', 'YearsAtCompany']]
X = normalize(X)
y = df['Yes'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors=5)
knn = knn.fit(X_train, y_train) 
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test) 
cm = confusion_matrix(y_test, y_test_pred) 
cm
```

```python
conf_matrix = pd.DataFrame(cm, index=['Yes', 'No'], columns = ['Attrition Yes', 'Attrition No']) 
conf_matrix
```
